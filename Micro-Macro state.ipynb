```python
import numpy as np
import matplotlib.pyplot as plt




#Aufgabe 2 Unterschied zwischen Mikrozustand und Makrozustand anhand von Münzwürfen
def fac(n):
    if n==1:
        return 1
    else:
        return n*fac(n-1)
    
def E(n,z):
    if z ==0 or z==n:
        return 1
    elif z==1 or z== n-1:
        return n
    else:
        E_n = n/fac(z)
        for i in range (1,z+1):
            E_n = E_n*(n-i)
    return E_n

def plot_list(n):
    liste = np.zeros(n)
    for i in range(0,n):
        liste[i] = E(n,i)
    plt.scatter(range(0,n),liste)
    plt.title("Multiplizität des Makrozustands n = {}".format(n))
    plt.grid()
    plt.show()
```

```python
plot_list(10)
```

```python
plot_list(20)
```
