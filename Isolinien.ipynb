```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

x = np.linspace(-3.5,3.5)
y = np.linspace(-3.5,3.5)


x,y = np.meshgrid(x,y)
z = np.cos(x)+np.cos(y)

#contour() = linien, contourf() = flchen
cs = plt.contour(x,y,z,16)
plt.clabel(cs, inline = 1, fontsize = 10)
plt.xlabel("$k_x$")
plt.ylabel("$k_y$")
plt.title("A2: Isoenergien der ersten Brilloninzone" )
plt.show()
```

```
fig = plt.figure()
ax = fig.gca(projection = "3d")
ax.set_xlabel("$k_x$")
ax.set_ylabel("$k_y$")
ax.set_zlabel("$E$")
ax.plot_surface(x,y,z, alpha = 0.1, cmap = "coolwarm")
ax.contour(x,y,z,16, cmap = "coolwarm")
ax.set_title("3D-Ansicht der Isolinien")
plt.show()
```
